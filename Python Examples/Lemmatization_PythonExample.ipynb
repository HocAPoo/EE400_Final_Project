```python
#========LEMMATIZATION=========
%pip install spacy

#To run, this requires a panda dataframe called data_clean 
#which has data labelled normalized_data
#adjust dataframe and column label names as necessary to suit your data

# Imports
import spacy
nlp = spacy.load("en_core_web_sm")  # Load spaCyâ€™s English model

# lemmatization function
def lemm_data(data):

    words = str(data).split()  # Split into string words
    lemmatized_words = [] # Collect lemmas here

    for word in words:
        if word.isalpha():  # If the word is alphabetic
            temp = nlp(word)  # Run spaCy on this single word, now nlp (doc) object
            lemma = temp[0].lemma_  # Get lemma (as string) of the first (only) word
            lemmatized_words.append(lemma.lower())  # Add lowercase lemma

    return " ".join(lemmatized_words)

#=======INITIALIZE LEMMATIZATION=======
data_clean.loc[:, "lemmatized_data"] = data_clean["normalized_data"].apply(lemm_data)
```
