```python
#=======ELIMINATE NOISE======
%pip install contractions
#To run, this requires a panda dataframe called data_clean (which can be created using the Load Data code/sample
#data_clean has raw data labelled 'raw_data'
#adjust dataframe and column label names as necessary to suit your data

# Import library to expand contractions 
#(for example:  "don't" will be replaced with "do not")
import contractions

# Define a function to eliminate noise by expanding contractions and expanding contractions
def eliminate_noise(data):

    # Step 0.5: maunual fixes
    data = data.replace("don;t", "don't")
    data = data.replace("Don;t", "Don't")
    
    # Step 1: expand contractions
    data = contractions.fix(str(data))

    # Step 2: eliminate special characters
    kept_chars = []
    for char in data:
        if char.isalnum() or char.isspace(): # if character is alphanumeric or a space
            kept_chars.append(char)
    data = "".join(kept_chars) # Add all kept characters back into data line

    return data

#=======INITIALIZE NOISE ELIMINATION=======
# Remove noise from data_clean and add onto panda set (i.e. add noise_removed_data to data_clean)
data_clean.loc[:, "noise_removed_data"] = data_clean["raw_data"].apply(eliminate_noise)
```
