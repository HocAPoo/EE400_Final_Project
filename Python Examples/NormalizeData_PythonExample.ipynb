```python
#======NORMALIZING=========
%pip install wordcloud

#To run, this requires a panda dataframe called data_clean 
#which has a column labelled noise_removed_data
#adjust dataframe and column label names as necessary to suit your data

# Import libraries needed for removing stopwords and for creating a word cloud
from wordcloud import WordCloud, STOPWORDS
import string

# Define function to normalize data
def normalize(data):

    # Step 1: Convert data to lowercase
    text = str(data).lower()
    
    # Step 2: Remove any punctuation
    text = "".join(char if char not in string.punctuation else " " for char in text)
    
    # (Optionial) Set custom stop words
    CUSTOM_STOPWORDS = set(STOPWORDS) - {"not"}  # eliminates "not" from the stopword list

    # Step 3: Remove stop words
    words = text.split() # Break text into individual words
    go_words = [] # Create an empty list to store the "good" words

    for word in words: # Loop through each word
        if word not in CUSTOM_STOPWORDS:
            go_words.append(word)
    
    return " ".join(go_words) # processed characters back into one string

#=======INITIALIZE NORMALIZING=======
# Remove noise from noise_removed_data and put preprocessed data into "noramalized_daa" column in data_clean
data_clean.loc[:, "normalized_data"] = data_clean["noise_removed_data"].apply(normalize)
```
